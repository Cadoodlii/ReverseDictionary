```python
import matplotlib.pyplot as plt
import torch
import numpy as np

def display_tensor_as_image(image_tensor):
    # Detach the tensor from the computational graph if it requires grad
    # and move it to CPU if it's on the GPU
    image_np = image_tensor.detach().cpu().numpy()

    # PyTorch typically uses [C, H, W] format, Matplotlib expects [H, W, C]
    if image_np.shape[0] in [1, 3]:
        image_np = image_np.transpose((1, 2, 0)) # Permute to [H, W, C]

    # Handle single-channel grayscale images
    if image_np.shape[2] == 1:
        image_np = image_np.squeeze()

    # Ensure pixel values are in the correct range for display (e.g., 0-1 or 0-255)
    # If the tensor is float but values are 0-255, convert to uint8
    if image_np.dtype == np.float32 and image_np.max() > 1:
        image_np = image_np.astype(np.uint8)

    plt.imshow(image_np)
    plt.axis('off') # Turn off axis labels
    plt.show()

# Example usage (assuming 'your_tensor' is a PyTorch tensor)
# display_tensor_as_image(your_tensor)
```

```python
display_tensor_as_image(image_tensor)
```
